```python
def dataframe_division(df,l):
    divided=np.array_split(df,l)
    for i in range(len(divided)):
        items=divided[i]
        items=items.reset_index()
        items=items.drop('index',axis=1)
        divided[i]=items
    final_df=pd.concat([items for items in divided],axis=1)
    return final_df
```
